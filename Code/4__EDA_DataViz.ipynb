## Exploratory Data Analysis


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
# JupyterLite
from js import fetch
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(dataset_part_2_csv)
df.head(5)
```

```python
# Local
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv')
df.head(5)
```

First, let's try to see how the payload mass and flight number (indicating the continuous launch attempts) variables would affect the launch outcome.

We can plot out the <code>PayloadMass</code> vs. <code>FlightNumber</code> and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass also appears to be a factor; even with more massive payloads, the first stage often returns successfully.


```python
ax = sns.catplot(x="FlightNumber", y="PayloadMass", hue="Class", data=df, s=40, aspect=5)
plt.xticks(np.arange(0, 100, 20)-1, fontsize=22)
plt.yticks(np.arange(0, 20000, 4000), fontsize=22)
plt.xlabel("Flight Number", fontsize=28)
plt.ylabel("Payload Mass (kg)", fontsize=28)
plb.setp(ax._legend.get_texts(), fontsize=22)
plb.setp(ax._legend.get_title(), fontsize=26)
plt.show()
```

Next, let's drill down to each site visualize its detailed launch records.


### TASK 1: Visualize the relationship between Launch Site and Flight Number


Use the function <code>catplot</code> to plot <code>LaunchSite</code> vs.
<code>FlightNumber</code>, set the parameter <code>x</code> parameter to <code>FlightNumber</code>, set the <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>.


```python
# Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value
ax = sns.catplot(x="FlightNumber", y="LaunchSite", hue="Class", data=df, s=40, aspect=5)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.xlabel("Flight Number", fontsize=28)
plt.ylabel("Launch Site", fontsize=28)
plb.setp(ax._legend.get_texts(), fontsize=22)
plb.setp(ax._legend.get_title(), fontsize=26)
plt.show()
```

Now let's try to explain the patterns found in this plot.


### Insights:
- As demonstrated in the previous plot, successful first-stage landing becomes more probable as the launch number increases, meaning that there is an improvment trend over time. This is also visible in the plot above. Therefore, it is generally safe to assume that a new launch will have a higher probablity than its formers to land the first stage successfully.
- The CCAFS SLC 40 launch site hosts significantly more launches than the other two sites over the period, except for a time window around flights #30-40 where it hosts no launches.
- However, from this data it seems that VAFB SLC 4E and KSC LC 39A have higher success rates.

### TASK 2: Visualize the relationship between Launch Site and Payload Mass





We also want to observe if there is any relationship between launch site and their payload mass:


```python
# Plot a scatter point chart with x axis to be Payload Mass (kg) and y axis to be the launch site, and hue to be the class value
ax = sns.catplot(x="PayloadMass", y="LaunchSite", hue="Class", data=df, s=40, aspect=5)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.xlabel("Payload Mass (kg)", fontsize=28)
plt.ylabel("Launch Site", fontsize=28)
plb.setp(ax._legend.get_texts(), fontsize=22)
plb.setp(ax._legend.get_title(), fontsize=26)
plt.show()
```

Now if you observe <code>LaunchSite</code> vs. <code>PayloadMass</code> scatter chart you will find for the VAFB SLC 4E launch site there are no rockets launched for heavy payload masses (greater than 10000kg).


### Insights:
- Most flights launched with a payload mass which is smaller than ~7000kg.
- No flights launched from the site CCAFS SLC 40 with payload mass between ~7500kg and ~14000kg.
- 100% success rate from site KSC LC 39A for payload mass < ~5500kg.
- Larger payloads generally generate higher rates of success.


### TASK 3: Visualize the relationship between Success Rate and Orbit Type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit:


```python
# HINT: use groupby method on Orbit column and get the mean of Class column
sns.barplot(x="Orbit", y="Class", data=df) #OR: df.groupby("Orbit").mean()['Class'].plot(kind='bar')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Orbit Type", fontsize=20)
plt.ylabel("Success Rate", fontsize=20)
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.


### Insights:
- High success orbits are: ES-L1, SSO, HEO, GEO, VLEO
- Except for the SO orbit which had no successful launches, all other orbits (that aren't mentioned above) have similar success rates.

### TASK 4: Visualize the relationship between Orbit Type and Flight Number


We want to see if there is any relationship between orbit type and flight number:


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
ax = sns.catplot(x="FlightNumber", y="Orbit", hue="Class", data=df, s=40, aspect=2)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Flight Number", fontsize=22)
plt.ylabel("Orbit Type", fontsize=22)
plb.setp(ax._legend.get_texts(), fontsize=16)
plb.setp(ax._legend.get_title(), fontsize=20)
plt.show()
```

You can observe that in the LEO orbit, success seems to be related to the number of flights. Conversely, in the GTO orbit, there appears to be no relationship between flight number and success.


### TASK  5: Visualize the relationship between Orbit Type and Payload Mass


Similarly, we can plot the <code>Orbit</code> vs. <code>PayloadMass</code>  scatter point charts to reveal the relationship between orbit type and payload mass:


```python
# Plot a scatter point chart with x axis to be Payload Mass and y axis to be the Orbit, and hue to be the class value
ax = sns.catplot(x="PayloadMass", y="Orbit", hue="Class", data=df, s=40, aspect=2)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Payload Mass (kg)", fontsize=22)
plt.ylabel("Orbit Type", fontsize=22)
plb.setp(ax._legend.get_texts(), fontsize=16)
plb.setp(ax._legend.get_title(), fontsize=20)
plt.show()
```

With heavy payloads the successful landing or positive landing rate are more for Polar, LEO and ISS.

However, for GTO, it's difficult to distinguish between successful and unsuccessful landings as both outcomes are present.


### TASK  6: Visualize the launch success yearly trend


We can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help getting the year from the date:


```python
# A function to Extract years from the date
year = []
def extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year

extract_year()
df['Date'] = year
df.head()
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
sns.lineplot(x="Date", y="Class", data=df)
plt.gcf().set_size_inches(12, 4)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel("Year", fontsize=20)
plt.ylabel("Success Rate", fontsize=20)
plt.show()
```

You can observe that the sucess rate since 2013 kept increasing till 2020.
